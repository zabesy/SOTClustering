Generate a word frequency bar plot after cleaning the data.

```python
total_word_count_dict = {}
school_word_count_dict = {}
for index, row in dataframe_sot.iterrows():
    if school_word_count_dict.get(row["school"]) is None:
        school_word_count_dict[row["school"]] = {}
    for word in row["tokenized_sentence"]:
        if word not in total_word_count_dict:
            total_word_count_dict[word] = 1
        else:
            total_word_count_dict[word] += 1
        if word not in school_word_count_dict[row["school"]]:
            school_word_count_dict[row["school"]][word] = 1
        else:
            school_word_count_dict[row["school"]][word] += 1
```

```python
#Sort the dictionaries into list of tuples
sorted_total_word_count = sorted(total_word_count_dict.items(), key=lambda x: x[1], reverse=True)

sorted_school_word_count = dict.fromkeys(dataframe_sot["school"].unique(), None)
for school, temp_dict in school_word_count_dict.items():
    sorted_school_word_count[school] = sorted(temp_dict.items(), key=lambda x: x[1], reverse=True)
```

```python
#Plot total word count.
count = 15
plt.bar(range(len(sorted_total_word_count[:count])), [value[1] for value in sorted_total_word_count[:count]], align='center')
plt.xticks(range(len(sorted_total_word_count[:count])), [val[0] for val in sorted_total_word_count[:count]])
plt.xticks(rotation=70)
plt.xlabel("Word")
plt.ylabel("Count")
plt.title(f"Total word frequency")
plt.show()
```

```python
#Plot word count per SOT.
for school, list_tuples in sorted_school_word_count.items():
    plt.figure(figsize=(8,6))
    plt.tight_layout()
    plt.bar(range(len(list_tuples[:count])), [value[1] for value in list_tuples[:count]], align='center')
    plt.xticks(range(len(list_tuples[:count])), [val[0] for val in list_tuples[:count]])
    plt.xticks(rotation=70)
    plt.xlabel("Word")
    plt.ylabel("Count")
    plt.title(f"{school.capitalize()}")
    plt.show()
```
